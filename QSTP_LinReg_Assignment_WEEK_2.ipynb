```python
import numpy as np
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
url='https://raw.githubusercontent.com/gurbaaz19/DeepLearningQSTP/main/QSTP_LinReg.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
fig,axs=plt.subplots(1,4,sharey=True)
df.plot(kind='scatter',x='AT',y='PE',ax=axs[0], figsize=(18,8))
df.plot(kind='scatter',x='V',y='PE',ax=axs[1])
df.plot(kind='scatter',x='AP',y='PE',ax=axs[2])
df.plot(kind='scatter',x='RH',y='PE',ax=axs[3])
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
lr.fit(X_train,y_train)
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1, normalize=False)
```

```python
y_pred = lr.predict(X_test)
plt.style.use('ggplot')
plt.plot(X_test,y_test,'x')
plt.plot(X_test,y_pred,'o')
```

```python
corr_matrix=np.corrcoef(y_test,y_pred)
```

```python
corr=corr_matrix[0,1]
```

```python
R_sq=pow(corr,2)
```
